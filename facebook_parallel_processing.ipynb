```python
def date_func(dt):
    now=datetime.now()
    print("input in date_func",dt)
    #monthDict
    monthDict = {'January':1,'February':2,'March':3,'April':4,'May':5,'June':6, 
                'July':7,'August':8,'September':9,'October':10,'November':11,'December':12}  
    try:
        if(dt == None):
            dt_= 'Not found'
        else:
            lst=dt.split()
            print('lst in date_func',lst)
        if (len(lst)==2):
            if (lst[1]=='m' or lst[1]=='h'):
                dt_= now.strftime('%d/%m/%Y')
            elif (lst[1]=='d'):
                delta_date = now - timedelta(days =int(lst[0]))
                dt_= delta_date.strftime('%d/%m/%Y')
            elif(lst[1]=='w'):
                print("inside elif length 2 with date in weeks")
                delta_date = now - timedelta(weeks =int(lst[0]))
                dt_= delta_date.strftime('%d/%m/%Y')
            else:
                print("inside else length 2") #27 April
                print(lst)
                lst.append('2020')
                print(monthDict[lst[1]])
                mm = monthDict[lst[1]]
                dt_=str(lst[0])+"/"+ str(mm) + "/"+ str(lst[2])
                print(dt_)
                #['29', 'December', '2020', 'at', '06:00']

        elif(len(lst)==3):
            print('length of list is 3')
            dt_=str(lst[0])+"/"+ str(monthDict[lst[1]]) + "/"+ str(lst[2])

        elif(len(lst)==4):
            lst[2]=lst[2]+lst[3]
            lst[2]='2020'
            dt_=str(lst[0])+"/"+ str(monthDict[lst[1]]) + "/"+ str(lst[2])
        else:
            dt_='Latest Activity Date Not Found'
    except Exception as e:
        pass
        #dt_="New date format"
    print('last_acivity_date_from_date_func',dt_)   
    return dt_
```
