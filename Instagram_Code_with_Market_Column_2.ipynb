# Instagram Data Scrapping

```python
def get_redirected_url(driver1):
    try:
        time.sleep(10)
        url_element = driver1.find_element(By.XPATH,"//a[contains(@page_id,'profilePage')]")
        desc_url = url_element.get_attribute("href")
        print("inside try of profilePage")
        desc_url = requests.get(urllib.parse.unquote(desc_url.split("?u=")[1].split("&e")[0])).url
        time.sleep(10)
        print("desc_url after redirection",desc_url)
    except Exception as e:
        desc_url=''
        #processed_desc_url=''
        print("desc_url is not present",desc_url)
    return desc_url,url_element
```

```python
def check_handler_verification(name,data_dictionary):
    try:
        if (name.find('Verified')!=-1):
            data_dictionary['Verified or Non Verified']='Verified'
        else:
            data_dictionary['Verified or Non Verified']='Non Verified'
    except Exception as e:
        print("url name parameter is missing")
    return data_dictionary   
```

```python
def get_latest_activity_timestamp(driver1,data_dictionary):
    try:
        time.sleep(10)
        lst = driver1.find_elements(By.XPATH,"//div[contains(@class,'Nnq7C weEfm')]//a")
        print("length of lst",len(lst))
        for item in lst[:3]:
            link_= item.get_attribute("href")
            #link=f"https://www.instagram.com{link}"
            print("Inside for loop of lst items",link_)
            driver1.get(link_)
            time.sleep(10)
            dt=driver1.find_element_by_tag_name("time")
            print(dt)
            dt_timestamp= dt.get_attribute('datetime')
            print('dt_timestamp',dt_timestamp)
            if dt_timestamp:
                print("Inside timestamp if")
                d = dateutil.parser.parse(dt_timestamp)
                dt= d.strftime('%d/%m/%Y')
                print("date",dt)
                data_dictionary['Latest Activity Date']=dt
                break;
            else:
                continue
    except Exception as e:
        print("latest_activity_date not present")
        data_dictionary['Latest Activity Date']="latest_activity_date not present"
    return data_dictionary
```

```python
def get_web_driver_elements_for_processing_keyword(key,driver):
    try:
        print("Inside of get_web_driver_elements_for_processing_keyword")
#         seg = dict_keywords_segs[key]
#         print(seg)
        searchbox=driver.find_element(By.XPATH,"//input[@placeholder='Search']");
        time.sleep(10)
        print("Inside searchbox of get_web_driver_elements_for_processing_keyword",searchbox)
        searchbox.send_keys(key)
        #searchbox.send_keys(key, Keys.ENTER)
        time.sleep(10)
        elements = driver.find_elements(By.XPATH,"//a[contains(@class,'-qQT3')]")
        time.sleep(10)
        #print("Inside elements of get_web_driver_elements_for_processing_keyword",elements)
    except Exception as e:
        print("Web_driver_elements_for_processing_keyword not found or Instagram account temporary locked")
    return elements,searchbox
```

```python
def get_processed_details_from_webdriver_element(index,driver1):
    try:
        name=index.text
        link=index.get_attribute("href")
        driver1.get(link)
        time.sleep(10)
        print('processing for %s handler'% link)
    except Exception as e:
        print("processed_details_from_webdriver_element not found")
    return name,link
```

```python
def get_associated_market(dict_keywords_segs):
    dict_keywords_seg_market = {}
    for i in range(len(dict_keywords_segs)):
        dict_keywords_seg_market[dict_keywords_segs['Keyword'][i]] = [dict_keywords_segs['Segment'][i],dict_keywords_segs['Market'][i]]
    return dict_keywords_seg_market
```

```python
#xpath for search on instagram
start=datetime.now()
list_data_dictionary=[]
count=0

dict_keywords_seg_market= get_associated_market(dict_keywords_segs)

print("dict_keywords_seg_market",dict_keywords_seg_market)

for key in dict_keywords_seg_market:
    try:
        count=count+1
        time.sleep(20)
        seg = dict_keywords_seg_market[key][0]
        print('Segemnt',seg)
        mar = dict_keywords_seg_market[key][1]
        print('Market',mar)
        print('Serial Number of keyword in keywords list',count)
        print('executing for the keyword ',key)
        elements,searchbox = get_web_driver_elements_for_processing_keyword(key,driver)
        for index in elements[:10]:
            try:
                name,link=get_processed_details_from_webdriver_element(index,driver1)
                desc_url=''
                #processed_desc_url=''
                keyword_in_lowercase= key.lower().replace(' ','')
            except Exception as e:
                print("driver1 is not able to process link:",link)
            try:
                desc_url,url_element=get_redirected_url(driver1)
                print('desc_url from get_redirected_url(driver1)',desc_url)
            except Exception as e:
                print("Redirected description url is not present or issues in redirection")     
            desc_url_in_lowercase= desc_url.lower().replace(' ','') 
            if (name.lower().find('verified')!=-1 or desc_url_in_lowercase.find(keyword_in_lowercase)!=-1 or desc_url_in_lowercase.find('mars.com')!=-1):
                try:
                    print("Inside try of either verified or handler containing descr_url")
                    data_dictionary={}
                    data_dictionary = check_handler_verification(name,data_dictionary)            
                    if(desc_url):
                        data_dictionary['Redirected_Desc_Url']=desc_url
                    else:
                        data_dictionary['Redirected_Desc_Url']=''                   
                    if(url_element.text):
                        data_dictionary['Text_Desc_Url'] = url_element.text
                        desc_txt=url_element.text
                        data_dictionary['Text_Desc_Url']=desc_txt       
                    else:
                        data_dictionary['Text_Desc_Url'] =''
            
                    #data_dictionary = get_txt_description_details(driver1,data_dictionary)
                    
                    data_dictionary = get_latest_activity_timestamp(driver1,data_dictionary)
                    
                    data_dictionary['Keyword'] = key
                    
                    data_dictionary['Segment'] = seg
                    
                    data_dictionary['Market'] = mar
                    
                    handler_name= name.split()[0]
                    
                    data_dictionary['Handler Name']= handler_name
                    
                    print('data_dictionary after handler name addition',data_dictionary)
                    
                    data_dictionary['Handler Url'] = link.replace("https://www.","")
                    
                    handler_url = data_dictionary['Handler Url']
                    
                    data_dictionary['Handler Url'] = handler_url[:-1]
                    
                   # print('data_dictionary after handler url',data_dictionary)
                    
                    list_data_dictionary.append(data_dictionary)
                    
                    print('list of data_dictionary after handler url',list_data_dictionary)
                    
                except Exception as e:
                    print("Exception")
                    continue
            else:
                print('%s is not an authentic handler url' % (link))
                continue
        searchbox.clear()
        time.sleep(20)
    except Exception as e:
        print("Something went wrong")
end = datetime.now()
driver.get("https://www.instagram.com/")
driver1.get("https://www.instagram.com/")
# print('*************Time taken: ',end-start,'****************')
```
